# Real-time simulation outputs and parallelization

In a robotics context we will want to be piping numbers into and out of simulations at a fairly rapid rate, interacting with sensors, motors, etc. Below are some scribbles on how to achieve this. 

We define a class that contains a simulation object and can run it as a background job using IPython's parallization engine. The class has a number of methods for querying the simulation outputs, a real-time plotting function that adds the latest simulation outputs as soon as they are produced. Getting real-time movie streaming working in the notebook requires a few extra things but is worth the effort. 

There's plenty of room for improvement here, such as  using an hdf5 storage container (see e.g. [here](http://stackoverflow.com/questions/14262433/large-data-work-flows-using-pandas)), running with gpu, etc. Any suggestions for improvements, please fire away.

## A run-in-the-backround simulator class

*Importage*

```python
import time
from tvb.simulator.lab import *

from IPython.lib import backgroundjobs as bg
from IPython import display

from copy import deepcopy
import datetime
import sys
import pandas as pd

import seaborn as sns

from matplotlib import pyplot as plt
%matplotlib inline
```

*Define the BackgroundSim class*

To use this class with different simulation parameters, provide a configured sim object at initialization  

```python
class BackgroundSim(object):
    
  """
  Usage:
  -----
  
  bgs = BackgrounSim()
  
  bgs.run_sim()
  bgs.run_sim_in_bg()
  bgs.make_graph()
  
  """
  def __init__(self, int_dt=2**-2, oscillator_params=None,sim=None):
        
    if not sim:  
      if not oscillator_params:            
        oscillator_params =  {'a':1.05,'b':-1.0,'c':0,'d':0.1,'e':0,'f':1/3,'g':1,
                              'I':0,'alpha':1,'beta':0.2,'gamma':-1,'tau':1.25}    
    oscillator = models.Generic2dOscillator()
    white_matter = connectivity.Connectivity(load_default=True)
    white_matter.speed = numpy.array([8.0])
    white_matter_coupling = coupling.Linear(a=0.0152)
    my_seed = 13
    my_random_state = numpy.random.RandomState(my_seed)
    hiss = noise.Additive(nsig=0.08)
    heunint = integrators.HeunStochastic(dt=int_dt, noise=hiss)
    heunint.configure()
    mama = monitors.TemporalAverage(period=2 ** -1)
    mimi = monitors.Bold()
    what_to_watch = (mama,mimi) #(momo, mama, mimi)
    sim = simulator.Simulator(model=oscillator, connectivity=white_matter,
                              coupling=white_matter_coupling,
                             integrator=heunint, monitors=what_to_watch)
    sim.configure()
    self.sim = sim
    self.rls = sim.connectivity.region_labels    
    self.bold_time = []  
    self.bold_data = []    
    self.tavg_time = []  
    self.tavg_data = []    
    self.jobs = bg.BackgroundJobManager()


  def run_sim(self,sim_length=10):
    self.sim_running=True
    for tavg,bold in self.sim(simulation_length=sim_length):
      if bold is not None:
        self.bold_time.append(bold[0])
        self.bold_data.append(bold[1])
      if tavg is not None:
        self.tavg_time.append(tavg[0])
        self.tavg_data.append(tavg[1])
    self.running=False
    
  def run_sim_in_bg(self,sim_length=10):
    self.jobs.new('self.run_sim(%s)' %(sim_length))   
    
  def list_arr_size(self,interval=1, reps=5):
    for n in range(reps):
      time.sleep(interval)
            
      display.clear_output(wait=True)
      display.display(plt.gcf())    
      sys.stdout.flush()
    
  def get_avg_arrs(self,varname):
    if varname=='tavg':
      tl = self.tavg_time
      ta = np.squeeze(np.array(self.tavg_time))[:len(tl)]
      da = np.squeeze(np.array(self.tavg_data))[:len(tl),:]
    elif varname=='bold':
      tl = self.bold_time
      if len(tl)>0:
        ta = np.squeeze(np.array(self.bold_time))[:len(tl)]
        da = np.squeeze(np.array(self.bold_data))[:len(tl),:]
      else:
        ta,da = np.nan,np.nan
    
    return ta,da


  def clock_time_report(self,sim_params,run_in_bg=False):
    import time
    start = time.time()
    if run_in_bg: self.run_sim_in_bg(**sim_params)
    else:         self.run_sim(**sim_params)
    finish = time.time()
    duration = (finish-start)*1000
    clock_ratio = sim_params['sim_length']/duration
    return duration, clock_ratio


              
  def make_graph(self,interval=1, reps=5,xlim=None,ylim=None,make_movie_frames=False,plot_params=None,
                 plot_tavg=True,plot_bold=False,vline_lims = [-1,1],
                 roi_nums=[0,10,20,30,40,50,60]):
    import time
    import numpy as np
    for n in range(reps):
      time.sleep(interval)
      for doplot,name in zip([plot_tavg,plot_bold],['tavg', 'bold']):
        if doplot:
          t,d = self.get_avg_arrs(name)
          if d is not np.nan:

            df = pd.DataFrame(d,index=t)[roi_nums]
            fig, ax = plt.subplots()            
            ax.axvline(x=t[-1], ymin=vline_lims[0],ymax=vline_lims[1],
                        linewidth=2, color='k', axes=ax,linestyle='dashed')
            ax = df.plot(title='Sim state at %s second(s)' %(n+1), ax=ax,
                         **plot_params);
            
      display.clear_output(wait=True)
      display.display(plt.gcf())    
      sys.stdout.flush()
      plt.close()
```

## Clock timing

Some notes on how long simulations take.

'Ratio' here is the ratio sim length / sim duration ; so values <1 mean it's slower than real-time, and >1 mean it's faster. 

```python
bgs2 = BackgroundSim()
duration,ratio = bgs2.clock_time_report(sim_params={'sim_length': 10})
```

```python
duration
```

```python
ratio
```

Now do a longer sim. 

```python
bgs3 = BackgroundSim()
duration,ratio = bgs3.clock_time_report(sim_params={'sim_length': 20000})
```

```python
duration
```

```python
ratio
```

The ratio is the same for the longer sim. So it is the simulator rather than the other code that is taking the time. Good. 


Try to speed it up with a longer integration time step:


```python
bgs4 = BackgroundSim(int_dt=1)#2**-2)
duration,ratio = bgs4.clock_time_report(sim_params={'sim_length': 10})
```
